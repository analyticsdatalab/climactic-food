```python
import os
import glob
import pandas as pd

from matplotlib import pyplot as plt
```

# Importing Data

### INDIA

```python
extension = 'csv'
india_all_filenames = [i for i in glob.glob('../data/india/*.csv')]    
```

```python
#combine all files in the list
india_combined_csv = pd.concat([pd.read_csv(f) for f in india_all_filenames ])
#export to csv
india_combined_csv.to_csv( "combined_india.csv", index=False, encoding='utf-8-sig')
```

```python
df_india = pd.read_csv("combined_india.csv")
```

## 5 Year Analysis

```python
df_india_time_range_0813 = df_temp_india.loc[(df_temp_india['date_time'] >= '2008-08-10') & (df_temp_india['date_time'] < '2013-08-11')]
```

```python
x_india_5y = df_india_time_range_0813['date_time']
y_india_5y = df_india_time_range_0813['tempC']

plt.figure(figsize=(15,5))
plt.scatter(x_india_5y, y_india_5y)

plt.show()
```

```python
df_china_time_range_0813 = df_temp_china.loc[(df_temp_china['date_time'] >= '2008-08-10') & (df_temp_china['date_time'] < '2013-08-11')]
```

```python
x_china_5y = df_china_time_range_0813['date_time']
y_china_5y = df_china_time_range_0813['tempC']

plt.figure(figsize=(15,5))
plt.scatter(x_china_5y, y_china_5y)

plt.show()
```

```python
df_brazil_time_range_08_fin = df_temp_brazil.loc[(df_temp_brazil['date_time'] >= '2008-08-10') & (df_temp_brazil['date_time'] < '2008-12-30')]
```

```python
x_brazil_5y = df_brazil_time_range_08_fin['date_time']
y_brazil_5y = df_brazil_time_range_08_fin['tempC']

plt.figure(figsize=(15,5))
plt.scatter(x_brazil_5y, y_brazil_5y)

plt.show()
```

```python
df_usa_time_range_0813 = df_temp_usa.loc[(df_temp_usa['date_time'] >= '2008-08-10') & (df_temp_usa['date_time'] < '2013-08-11')]
```

```python
x_usa_5y = df_usa_time_range_0813['date_time']
y_usa_5y = df_usa_time_range_0813['tempC']

plt.figure(figsize=(15,5))
plt.scatter(x_usa_5y, y_usa_5y)

plt.show()
```

```python
fig, axs = plt.subplots(4)
fig.suptitle('Weather Data Analysis 5 Years Comparison')
plt.subplots_adjust(bottom=0.1)
axs[0].set_title('USA')
axs[0].plot(x_usa_5y, y_usa_5y)
axs[1].set_title('Brazil')
axs[1].plot(x_brazil_5y, y_brazil_5y, 'tab:orange')
axs[2].set_title('China')
axs[2].plot(x_china_5y, y_china_5y, 'tab:green')
axs[3].set_title('India')
axs[3].plot(x_india_5y, y_india_5y, 'tab:red')
```
