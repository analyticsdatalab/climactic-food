```python
import pandas as pd
import os

from matplotlib import pyplot as plt
```

## Americana, Brasil Data Analysis

```python
file_name = "americana.csv"
folder_path = "../../../WorldWeatherOnline/data"
full_file_path = os.path.join(folder_path, file_name)
```

```python
df = pd.read_csv(full_file_path)
```

```python
df_temp = df[['date_time', 'maxtempC', 'mintempC', 'tempC', 'precipMM']]
```

```python
df_temp.head(10)
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(df.date_time, df.tempC, color="red", marker="o")
# set x-axis label
ax.set_xlabel("date_time",fontsize=14)
# set y-axis label
ax.set_ylabel("tempC",color="red",fontsize=14)
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df.date_time, df.precipMM,color="blue",marker="o")
ax2.set_ylabel("precipMM",color="blue",fontsize=14)
plt.show()
```

```python
# save the plot as a file
fig.savefig('precipitations_temperatures.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
