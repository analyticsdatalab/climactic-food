```python
import pandas as pd
import os
import glob

from matplotlib import pyplot as plt
```

```python
extension = 'csv'
all_filenames_romania = [i for i in glob.glob('../data/romania/*.csv')]    
```

```python
#combine all files in the list
combined_csv_romania = pd.concat([pd.read_csv(f) for f in all_filenames_romania ])
#export to csv
combined_csv_romania.to_csv( "combined_romania.csv", index=False, encoding='utf-8-sig')
```

```python
df_romania = pd.read_csv("combined_romania.csv")
```

# Descriptive Data

```python
df_romania.head(10)
```

```python
df_romania.tail()
```

```python
df_romania.describe()
```

```python
df_romania.info()
```

```python
df_temp_romania = df_romania[['date_time', 'tempC', 'precipMM']]
```

```python
df_temp_romania.head()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()

# make a plot
ax.plot(df_romania.date_time, df_romania.tempC, color="red", marker="o")

# set x-axis label
ax.set_xlabel("date_time",fontsize=14)

# set y-axis label
ax.set_ylabel("tempC",color="red",fontsize=14)
ax2=ax.twinx()

# make a plot with different y-axis using second axis object
ax2.plot(df_romania.date_time, df_romania.precipMM,color="blue",marker="x")
ax2.set_ylabel("precipMM",color="blue",fontsize=14)
plt.show()
```

```python
# save the plot as a file
fig.savefig('precipitations_temperatures_romania_all.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
