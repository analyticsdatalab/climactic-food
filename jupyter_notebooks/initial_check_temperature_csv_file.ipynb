```python
import pandas as pd
import os

# importing matplotlib module  
from matplotlib import pyplot as plt  
```

```python
file_name = "temperature.csv"
folder_path = "../data"
full_file_path = os.path.join(folder_path, file_name)
```

```python
df = pd.read_csv(full_file_path)
```

```python
df['Date'] = pd.to_datetime([f'{y}-{m}-{d}' for y, m, d in zip(df.year, df.month, df.day)])
```

```python
df.head()
```

```python
df_temp = df[['Date','Country', 'City', 'AverageTemperatureFahr', 'year']]
```

```python
df_temp
```

```python
df_plot =  df_temp[(df_temp['Country'] == 'Brazil') & (df_temp['City'] == 'Brasília') & (df_temp['year'] >= 2000)]
df_plot.head()
```

```python
y = df_plot['AverageTemperatureFahr']
x = df_plot['Date']

plt.figure(figsize=(15,4))
plt.scatter(x, y) 
  
# function to show the plot  
plt.show() 
```
