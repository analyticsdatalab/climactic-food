```python
import pandas as pd
import os
import glob

from matplotlib import pyplot as plt
```

```python
extension = 'csv'
all_filenames_france = [i for i in glob.glob('../data/france/*.csv')]  
```

```python
#combine all files in the list
combined_csv_france = pd.concat([pd.read_csv(f) for f in all_filenames_france ])

#export to csv
combined_csv_france.to_csv( "combined_france.csv", index=False, encoding='utf-8-sig')
```

```python
df_france = pd.read_csv("combined_france.csv")
```

# Descriptive Data

```python
df_france.head(10)
```

```python
df_france.tail()
```

```python
df_france.describe()
```

```python
df_france.info()
```

```python
df_temp_france = df_france[['date_time', 'tempC', 'precipMM']]
```

```python
df_temp_france.head()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()

# make a plot
ax.plot(df_france.date_time, df_france.tempC, color="red", marker="o")

# set x-axis label
ax.set_xlabel("date_time",fontsize=14)

# set y-axis label
ax.set_ylabel("tempC",color="red",fontsize=14)
ax2=ax.twinx()

# make a plot with different y-axis using second axis object
ax2.plot(df_france.date_time, df_france.precipMM,color="blue",marker="x")
ax2.set_ylabel("precipMM",color="blue",fontsize=14)
plt.show()
```

```python
# save the plot as a file
fig.savefig('precipitations_temperatures_france_all.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
